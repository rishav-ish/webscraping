# Image Retriever on WTFFunFacts Site

```python
import os
import requests
from bs4 import BeautifulSoup
```

```python
path = os.getcwd() 
```

# Requesting site 

```python
url = "https://wtffunfact.com/"

html_response = requests.get(url)
```

# Beautifying it

```python
bsObj = BeautifulSoup(html_response.content)
```

```python
bsObj
```

# Retrieving all image tag

```python
img_tag = bsObj.find_all("img")              #for look-up
```

```python
url_to_retrieve = set()

for img in img_tag:
    if img['src'].startswith("http") not in url_to_retrieve:
        url_to_retrieve.add(img['src'])
```

```python
url_to_retrieve = list(url_to_retrieve)        #for indexing 
```

```python
len(url_to_retrieve)
```

# Downloading these images

```python
os.mkdir(path + "//image")
```

```python
path = path + "//image"
```
